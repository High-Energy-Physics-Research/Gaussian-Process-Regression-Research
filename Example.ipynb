```matlab
  x = gpml_randn(0.8, 20, 1);                 % 20 training inputs
  y = sin(3*x) + 0.1*gpml_randn(0.9, 20, 1);  % 20 noisy training targets
  xs = linspace(-3, 3, 61)';                  % 61 test inputs 
```

```matlab
figure;
hold on

% dados de treino
plot(x, y, 'kx', 'MarkerSize', 8, 'LineWidth', 1.5)

% funo verdadeira (sem rudo)
plot(xs, sin(3*xs), 'b-', 'LineWidth', 1.5)

xlabel('x')
ylabel('y')
legend('training data', 'true function', 'Location', 'best')
grid on
box on

hold off
```
