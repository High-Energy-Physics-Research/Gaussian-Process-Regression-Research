# Hidrodinâmica Relativística como Teoria Efetiva

A hidrodinâmica relativística descreve a evolução de sistemas quânticos de muitos corpos
em termos de campos coletivos, como a densidade de energia e o campo de velocidades.
Esses campos não são fundamentais, mas emergem após um processo de coarse-graining
dos graus de liberdade microscópicos.
Nesse regime, a validade da descrição exige equilíbio local e gradientes pequenos.
A dinâmica resultante é governada por leis de conservação,
que codificam informações microscópicas de forma altamente comprimida.
Assim, a hidrodinâmica pode ser vista como uma teoria efetiva universal,
independente dos detalhes da interação fundamental.
Essa universalidade é uma das razões de seu enorme poder descritivo.

A conservação do tensor energia-momento é expressa por
$$
\partial_\mu T^{\mu\nu} = 0.
$$

Para um fluido perfeito, o tensor energia-momento assume a forma
$$
T^{\mu\nu} = (\varepsilon + p) u^\mu u^\nu - p\, g^{\mu\nu},
$$
onde $\varepsilon$ é a densidade de energia, $p$ a pressão e $u^\mu u_\mu = 1$.

No caso de expansão longitudinal simples, a equação de evolução da energia se reduz a
$$
\frac{d\varepsilon}{d\tau} = -\frac{\varepsilon + p}{\tau},
$$
mostrando explicitamente como a expansão do sistema controla o resfriamento do fluido.
